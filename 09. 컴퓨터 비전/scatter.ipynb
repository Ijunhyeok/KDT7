```python
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score

# 1. ë°ì´í„° ë¡œë“œ
df = pd.read_csv(r"./í•‘í¬_ì¤‘ì•™Crop_RGB_HSV_KMeans.csv")

from sklearn.cluster import KMeans
X_color = df[['H_mean', 'S_mean', 'V_mean']]
kmeans = KMeans(n_clusters=5, random_state=42)
df['color_cluster'] = kmeans.fit_predict(X_color)


# print(df['color_cluster'].value_counts())

# 3. íŠ¹ì„±ê³¼ íƒ€ê²Ÿ ë¶„ë¦¬
X = df[['R_mean', 'G_mean', 'B_mean', 'R_std', 'G_std', 'B_std', 'H_mean', 'S_mean', 'V_mean']]
y = df['color_cluster']  # í°ìƒ‰ ì—¬ë¶€ë¥¼ íƒ€ê²Ÿìœ¼ë¡œ ì„¤ì •

# 4. í•™ìŠµ/í…ŒìŠ¤íŠ¸ ë¶„ë¦¬
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 5. SVMìš© ì •ê·œí™”
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 6. ëª¨ë¸ ì •ì˜
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "LightGBM": LGBMClassifier(random_state=42)
}

# 7. í•™ìŠµ ë° í‰ê°€
print("í°ìƒ‰ vs ê¸°íƒ€ ë¶„ë¥˜ ê²°ê³¼:\n")
for name, model in models.items():
    if name == "SVM":
        model.fit(X_train_scaled, y_train)
        train_pred = model.predict(X_train_scaled)
        test_pred = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)

    train_score = accuracy_score(y_train, train_pred)
    test_score = accuracy_score(y_test, test_pred)

    print(f"ğŸ”¹ {name}")
    print(f"   - Train Accuracy: {train_score:.4f}")
    print(f"   - Test Accuracy:  {test_score:.4f}\n") 
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import joblib

# # ìƒ‰ìƒ ë¶„ë¥˜ í•¨ìˆ˜
# def classify_color_group_from_h(h_value):
#     if 0 <= h_value < 30:
#         return "Red"
#     elif 30 <= h_value < 90:
#         return "Orange/Yellow"
#     elif 90 <= h_value < 150:
#         return "Green"
#     elif 150 <= h_value < 210:
#         return "Blue"
#     elif 210 <= h_value < 270:
#         return "Purple"
#     else:
#         return "Others"

# KMeansë¡œ í´ëŸ¬ìŠ¤í„° ìƒì„±
def create_kmeans_clusters(df, features, n_clusters=5):
    X_color = df[features]
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df['color_cluster'] = kmeans.fit_predict(X_color)
    print("ğŸ”¹ KMeans í´ëŸ¬ìŠ¤í„°ë§ ì™„ë£Œ!")
    return kmeans

# Decision Tree í•™ìŠµ ë° êµì°¨ ê²€ì¦ í•¨ìˆ˜
def train_decision_tree(X, y):
    # ë°ì´í„° ë¶„ë¦¬
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

    # ëª¨ë¸ ì •ì˜
    model = DecisionTreeClassifier(random_state=42)

    # êµì°¨ ê²€ì¦ (5-í´ë“œ)
    print("ğŸ”¹ Cross-Validation Scores:")
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    print(f"   - CV Mean Accuracy: {np.mean(cv_scores):.4f}")
    print(f"   - CV Accuracy Scores: {cv_scores}\n")

    # ëª¨ë¸ í•™ìŠµ ë° ì˜ˆì¸¡
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # í˜¼ë™ í–‰ë ¬ ë° ì„±ëŠ¥ ë³´ê³ 
    print("ğŸ”¹ Confusion Matrix:")
    conf_matrix = confusion_matrix(y_test, y_pred)
    print(conf_matrix)

    print("\nğŸ”¹ Classification Report:")
    class_report = classification_report(y_test, y_pred)
    print(class_report)

    # í•™ìŠµ ì •í™•ë„ ì¶œë ¥
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    print(f"ğŸ”¹ Decision Tree Final Accuracy:")
    print(f"   - Train Accuracy: {accuracy_score(y_train, train_pred):.4f}")
    print(f"   - Test Accuracy: {accuracy_score(y_test, test_pred):.4f}\n")

    return model

# ê°ì • ë¶„ì„ í•¨ìˆ˜
def assign_emotion(features):
    R_mean, G_mean, B_mean, R_std, G_std, B_std, H_mean, S_mean, V_mean = features

    # ê°ì • ë¶„ë¥˜ ê¸°ì¤€ (í•‘í¬ ì¤‘ì‹¬ ìƒ‰ìƒê°’ ì¡°ì •)
    if R_mean >= 220 and G_mean >= 170 and B_mean >= 200:  # ì‚¬ë‘: ë°ê³  ë¶€ë“œëŸ¬ìš´ í•‘í¬í†¤
        return "ì‚¬ë‘(ë¼ì´íŠ¸ í•‘í¬)"
    elif R_std >= 210 and G_std >= 140 and B_std <= 160:  # í–‰ë³µ: ë”°ëœ»í•˜ê³  ìƒê¸°ìˆëŠ” í•‘í¬í†¤
        return "í–‰ë³µ(í•« í•‘í¬)"
    elif H_mean >= 20 and S_mean >= 40 and V_mean >= 70:  # í¬ë§: ë°ê³  í™œê¸°ì°¬ í•‘í¬í†¤
        return "í¬ë§(ë”¥ í•‘í¬)"
    elif 190 <= R_mean < 220 and 150 <= G_mean < 180 and 130 <= B_mean < 170:  # ìš°ì•„í•¨: ë¶€ë“œëŸ½ê³  ì„¸ë ¨ëœ í•‘í¬í†¤
        return "ìš°ì•„í•¨(í˜ì¼ í•‘í¬)"
    elif R_mean <= 160 and G_mean <= 150 and B_mean <= 130:  # í¸ì•ˆí•¨: ì–´ë‘ìš´ í•‘í¬í†¤
        return "í¸ì•ˆí•¨(ë”ìŠ¤í‹° ë¡œì¦ˆ)"
    elif 160 <= R_mean < 200 and 120 <= G_mean < 160 and 100 <= B_mean < 140:  # ìˆœìˆ˜í•¨: ì¤‘ê°„ í•‘í¬í†¤
        return "ìˆœìˆ˜í•¨(ë¼ì´íŠ¸ ì½”ë„)"
    elif H_mean <= 10 and S_mean <= 30 and V_mean <= 50:  # ê¸°íƒ€: ì–´ë‘ìš´ ìƒ‰ìƒ (í•‘í¬ ìŒì˜ í¬í•¨)
        return "ê¸°íƒ€(ìŠ¬ë ˆì´íŠ¸ ê·¸ë ˆì´)"
    else:
        return "ê¸°íƒ€(ìŠ¬ë ˆì´íŠ¸ ê·¸ë ˆì´)"

# ìƒˆë¡œìš´ ì´ë¯¸ì§€ ë¶„ë¥˜ ë° ê°ì • ì¶œë ¥ í•¨ìˆ˜
def classify_new_image(model, image_path):
    # ì´ë¯¸ì§€ íŠ¹ì§• ì¶”ì¶œ
    features = extract_features_from_image(image_path)
    features_array = np.array(features).reshape(1, -1)

    # Decision Tree ì˜ˆì¸¡
    dt_prediction = model.predict(features_array)[0]

    # ê°ì • ë¶„ì„
    emotion = assign_emotion(features)
    return dt_prediction, emotion

# íŠ¹ì§• ì¶”ì¶œ í•¨ìˆ˜
def extract_features_from_image(image_path):
    # ì´ë¯¸ì§€ ë¡œë“œ
    image = cv2.imread(image_path)
    if image is None:
        raise FileNotFoundError(f"ì´ë¯¸ì§€ë¥¼ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {image_path}")

    # RGB í‰ê·  ë° í‘œì¤€í¸ì°¨ ê³„ì‚°
    R_mean = np.mean(image[:, :, 2])
    G_mean = np.mean(image[:, :, 1])
    B_mean = np.mean(image[:, :, 0])
    R_std = np.std(image[:, :, 2])
    G_std = np.std(image[:, :, 1])
    B_std = np.std(image[:, :, 0])

    # HSV ë³€í™˜ ë° í‰ê·  ë° í‘œì¤€í¸ì°¨ ê³„ì‚°
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    H_mean = np.mean(hsv_image[:, :, 0])
    S_mean = np.mean(hsv_image[:, :, 1])
    V_mean = np.mean(hsv_image[:, :, 2])

    # íŠ¹ì§• ë°˜í™˜
    return [R_mean, G_mean, B_mean, R_std, G_std, B_std, H_mean, S_mean, V_mean]

# ê°ì • ë¶„ë¥˜ ë¼ë²¨ ìƒì„±
def assign_emotion_row(row):
    # í–‰ì˜ RGB ë° HSV íŠ¹ì§•ê°’ ì¶”ì¶œ
    features = [
        row['R_mean'], row['G_mean'], row['B_mean'], 
        row['R_std'], row['G_std'], row['B_std'], 
        row['H_mean'], row['S_mean'], row['V_mean']
    ]
    return assign_emotion(features)

# # ìƒˆë¡œìš´ ì»¬ëŸ¼ì— ê°ì • ë¶„ë¥˜ ê²°ê³¼ ì €ì¥
# df['emotion_group'] = df.apply(assign_emotion_row, axis=1)

# ê²°ê³¼ í™•ì¸
# print(df)

# ë©”ì¸ ì½”ë“œ
if __name__ == "__main__":
    # ë°ì´í„° ë¡œë“œ
    df = pd.read_csv(r"./í•‘í¬_ì¤‘ì•™Crop_RGB_HSV_KMeans.csv")
    
	# ê°ì • ë¶„ë¥˜ ë¼ë²¨ ìƒì„±
    df['emotion_group'] = df.apply(assign_emotion_row, axis=1)
    
    # íŠ¹ì§• ì •ì˜
    color_features = ['H_mean', 'S_mean', 'V_mean']
    full_features = ['R_mean', 'G_mean', 'B_mean', 'R_std', 'G_std', 'B_std', 'H_mean', 'S_mean', 'V_mean']
    target = 'emotion_group'

    # í”¼ì²˜ ë° íƒ€ê²Ÿ ì •ì˜
    X = df[full_features]
    y = df[target]

    # Label Encoding ì ìš©
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    # KMeans í´ëŸ¬ìŠ¤í„° ìƒì„±
    kmeans_model = create_kmeans_clusters(df, color_features)

    # Decision Tree ëª¨ë¸ í•™ìŠµ ë° ê²€ì¦
    decision_tree_model = train_decision_tree(X, y_encoded)

    # ëª¨ë¸ ì €ì¥
    joblib.dump({
        'model': decision_tree_model,
        'features': full_features  # ì •í™•í•œ ì»¬ëŸ¼ ìˆœì„œ ë¦¬ìŠ¤íŠ¸
    }, 'decision_tree_model.pkl')

    print("í•™ìŠµ ì‹œ ì‚¬ìš©ëœ feature ì»¬ëŸ¼:")
    print(X.columns.tolist())

    print("ğŸ”¹ ëª¨ë¸ ì €ì¥ ì™„ë£Œ (decision_tree_model.pkl)")

    # ìƒˆë¡œìš´ ì´ë¯¸ì§€ ë¶„ë¥˜ ë° ì¶œë ¥
    image_path = r"./test/07696733.jpeg"  # ì…ë ¥ ì´ë¯¸ì§€ íŒŒì¼ ê²½ë¡œ
    try:
        dt_pred, emotion = classify_new_image(decision_tree_model, image_path)
        print(f"ğŸ”¹ ìƒˆë¡œìš´ ì´ë¯¸ì§€ ì˜ˆì¸¡ ê²°ê³¼:")
        print(f"   - Decision Tree í´ëŸ¬ìŠ¤í„°: {dt_pred}")
        print(f"   - ê°ì • ë¶„ì„ ê²°ê³¼: {emotion}")
    except FileNotFoundError as e:
        print(e)
```
