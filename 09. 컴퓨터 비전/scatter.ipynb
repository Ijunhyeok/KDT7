```python
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score

# 1. 데이터 로드
df = pd.read_csv(r"./핑크_중앙Crop_RGB_HSV_KMeans.csv")

from sklearn.cluster import KMeans
X_color = df[['H_mean', 'S_mean', 'V_mean']]
kmeans = KMeans(n_clusters=5, random_state=42)
df['color_cluster'] = kmeans.fit_predict(X_color)


# print(df['color_cluster'].value_counts())

# 3. 특성과 타겟 분리
X = df[['R_mean', 'G_mean', 'B_mean', 'R_std', 'G_std', 'B_std', 'H_mean', 'S_mean', 'V_mean']]
y = df['color_cluster']  # 흰색 여부를 타겟으로 설정

# 4. 학습/테스트 분리
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 5. SVM용 정규화
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 6. 모델 정의
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "LightGBM": LGBMClassifier(random_state=42)
}

# 7. 학습 및 평가
print("흰색 vs 기타 분류 결과:\n")
for name, model in models.items():
    if name == "SVM":
        model.fit(X_train_scaled, y_train)
        train_pred = model.predict(X_train_scaled)
        test_pred = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)

    train_score = accuracy_score(y_train, train_pred)
    test_score = accuracy_score(y_test, test_pred)

    print(f"🔹 {name}")
    print(f"   - Train Accuracy: {train_score:.4f}")
    print(f"   - Test Accuracy:  {test_score:.4f}\n") 
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import joblib

# # 색상 분류 함수
# def classify_color_group_from_h(h_value):
#     if 0 <= h_value < 30:
#         return "Red"
#     elif 30 <= h_value < 90:
#         return "Orange/Yellow"
#     elif 90 <= h_value < 150:
#         return "Green"
#     elif 150 <= h_value < 210:
#         return "Blue"
#     elif 210 <= h_value < 270:
#         return "Purple"
#     else:
#         return "Others"

# KMeans로 클러스터 생성
def create_kmeans_clusters(df, features, n_clusters=5):
    X_color = df[features]
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df['color_cluster'] = kmeans.fit_predict(X_color)
    print("🔹 KMeans 클러스터링 완료!")
    return kmeans

# Decision Tree 학습 및 교차 검증 함수
def train_decision_tree(X, y):
    # 데이터 분리
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

    # 모델 정의
    model = DecisionTreeClassifier(random_state=42)

    # 교차 검증 (5-폴드)
    print("🔹 Cross-Validation Scores:")
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    print(f"   - CV Mean Accuracy: {np.mean(cv_scores):.4f}")
    print(f"   - CV Accuracy Scores: {cv_scores}\n")

    # 모델 학습 및 예측
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # 혼동 행렬 및 성능 보고
    print("🔹 Confusion Matrix:")
    conf_matrix = confusion_matrix(y_test, y_pred)
    print(conf_matrix)

    print("\n🔹 Classification Report:")
    class_report = classification_report(y_test, y_pred)
    print(class_report)

    # 학습 정확도 출력
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    print(f"🔹 Decision Tree Final Accuracy:")
    print(f"   - Train Accuracy: {accuracy_score(y_train, train_pred):.4f}")
    print(f"   - Test Accuracy: {accuracy_score(y_test, test_pred):.4f}\n")

    return model

# 감정 분석 함수
def assign_emotion(features):
    R_mean, G_mean, B_mean, R_std, G_std, B_std, H_mean, S_mean, V_mean = features

    # 감정 분류 기준 (핑크 중심 색상값 조정)
    if R_mean >= 220 and G_mean >= 170 and B_mean >= 200:  # 사랑: 밝고 부드러운 핑크톤
        return "사랑(라이트 핑크)"
    elif R_std >= 210 and G_std >= 140 and B_std <= 160:  # 행복: 따뜻하고 생기있는 핑크톤
        return "행복(핫 핑크)"
    elif H_mean >= 20 and S_mean >= 40 and V_mean >= 70:  # 희망: 밝고 활기찬 핑크톤
        return "희망(딥 핑크)"
    elif 190 <= R_mean < 220 and 150 <= G_mean < 180 and 130 <= B_mean < 170:  # 우아함: 부드럽고 세련된 핑크톤
        return "우아함(페일 핑크)"
    elif R_mean <= 160 and G_mean <= 150 and B_mean <= 130:  # 편안함: 어두운 핑크톤
        return "편안함(더스티 로즈)"
    elif 160 <= R_mean < 200 and 120 <= G_mean < 160 and 100 <= B_mean < 140:  # 순수함: 중간 핑크톤
        return "순수함(라이트 코랄)"
    elif H_mean <= 10 and S_mean <= 30 and V_mean <= 50:  # 기타: 어두운 색상 (핑크 음영 포함)
        return "기타(슬레이트 그레이)"
    else:
        return "기타(슬레이트 그레이)"

# 새로운 이미지 분류 및 감정 출력 함수
def classify_new_image(model, image_path):
    # 이미지 특징 추출
    features = extract_features_from_image(image_path)
    features_array = np.array(features).reshape(1, -1)

    # Decision Tree 예측
    dt_prediction = model.predict(features_array)[0]

    # 감정 분석
    emotion = assign_emotion(features)
    return dt_prediction, emotion

# 특징 추출 함수
def extract_features_from_image(image_path):
    # 이미지 로드
    image = cv2.imread(image_path)
    if image is None:
        raise FileNotFoundError(f"이미지를 불러올 수 없습니다: {image_path}")

    # RGB 평균 및 표준편차 계산
    R_mean = np.mean(image[:, :, 2])
    G_mean = np.mean(image[:, :, 1])
    B_mean = np.mean(image[:, :, 0])
    R_std = np.std(image[:, :, 2])
    G_std = np.std(image[:, :, 1])
    B_std = np.std(image[:, :, 0])

    # HSV 변환 및 평균 및 표준편차 계산
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    H_mean = np.mean(hsv_image[:, :, 0])
    S_mean = np.mean(hsv_image[:, :, 1])
    V_mean = np.mean(hsv_image[:, :, 2])

    # 특징 반환
    return [R_mean, G_mean, B_mean, R_std, G_std, B_std, H_mean, S_mean, V_mean]

# 감정 분류 라벨 생성
def assign_emotion_row(row):
    # 행의 RGB 및 HSV 특징값 추출
    features = [
        row['R_mean'], row['G_mean'], row['B_mean'], 
        row['R_std'], row['G_std'], row['B_std'], 
        row['H_mean'], row['S_mean'], row['V_mean']
    ]
    return assign_emotion(features)

# # 새로운 컬럼에 감정 분류 결과 저장
# df['emotion_group'] = df.apply(assign_emotion_row, axis=1)

# 결과 확인
# print(df)

# 메인 코드
if __name__ == "__main__":
    # 데이터 로드
    df = pd.read_csv(r"./핑크_중앙Crop_RGB_HSV_KMeans.csv")
    
	# 감정 분류 라벨 생성
    df['emotion_group'] = df.apply(assign_emotion_row, axis=1)
    
    # 특징 정의
    color_features = ['H_mean', 'S_mean', 'V_mean']
    full_features = ['R_mean', 'G_mean', 'B_mean', 'R_std', 'G_std', 'B_std', 'H_mean', 'S_mean', 'V_mean']
    target = 'emotion_group'

    # 피처 및 타겟 정의
    X = df[full_features]
    y = df[target]

    # Label Encoding 적용
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    # KMeans 클러스터 생성
    kmeans_model = create_kmeans_clusters(df, color_features)

    # Decision Tree 모델 학습 및 검증
    decision_tree_model = train_decision_tree(X, y_encoded)

    # 모델 저장
    joblib.dump({
        'model': decision_tree_model,
        'features': full_features  # 정확한 컬럼 순서 리스트
    }, 'decision_tree_model.pkl')

    print("학습 시 사용된 feature 컬럼:")
    print(X.columns.tolist())

    print("🔹 모델 저장 완료 (decision_tree_model.pkl)")

    # 새로운 이미지 분류 및 출력
    image_path = r"./test/07696733.jpeg"  # 입력 이미지 파일 경로
    try:
        dt_pred, emotion = classify_new_image(decision_tree_model, image_path)
        print(f"🔹 새로운 이미지 예측 결과:")
        print(f"   - Decision Tree 클러스터: {dt_pred}")
        print(f"   - 감정 분석 결과: {emotion}")
    except FileNotFoundError as e:
        print(e)
```
