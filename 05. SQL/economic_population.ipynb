```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
total_economic_population = pd.read_csv('economic_population.csv')
total_economic_population
```

```python
plt.figure(figsize=(10,5))
plt.plot(total_economic_population['year'], total_economic_population['20_29'], color='orange', label='20_29')
plt.plot(total_economic_population['year'], total_economic_population['30_39'], color='yellow', label='30_39')
plt.plot(total_economic_population['year'], total_economic_population['40_49'], color='green', label='40_49')
plt.plot(total_economic_population['year'], total_economic_population['50_59'], color='blue', label='50_59')
plt.plot(total_economic_population['year'], total_economic_population['60+'], color='navy', label='60+')


years = total_economic_population['year']
plt.xticks(years[::3])
plt.title('연도별, 연령별 경제활동인구(천명)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(total_economic_population['year'], total_economic_population['15+'], color='red', label='15+')
plt.plot(total_economic_population['year'], total_economic_population['15_64'], color='purple', label='15_64')
plt.title('연도별 경제활동 총인구 및 구역별 총인구(천명)')
years = total_economic_population['year']
plt.xticks(years[::3])
plt.legend()
plt.grid()
plt.show()
```

```python
BEESN = pd.read_csv('BEESN.csv')
BEESN
```

```python
plt.figure(figsize=(10,5))
plt.plot(BEESN['year'], BEESN['Business_count'], color='red', label='사업체 수')

years = BEESN['year']
plt.xticks(years[::2])
plt.title('연도별, 사업체 수(백만)')
plt.xlim(1998,2023)
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(BEESN['year'], BEESN['salaried_employees'], color='orange', label='임금근로자')

years = BEESN['year']
plt.xticks(years[::2])
plt.title('연도별, 임금근로자(천만명)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(BEESN['year'], BEESN['regular_employees'], color='yellow', label='상용근로자')
plt.plot(BEESN['year'], BEESN['temporary_or_daily_workers'], color='green', label='임시 및 일용근로자')

years = BEESN['year']
plt.xticks(years[::2])
plt.title('연도별, 임금근로자 형태(천만명)')
plt.legend()
plt.grid()
plt.show()
```
