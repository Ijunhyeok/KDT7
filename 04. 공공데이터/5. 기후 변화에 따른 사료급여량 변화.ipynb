```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import matplotlib.ticker as ticker
import numpy as np
```

```python
전국평균기온 = pd.read_excel('전국평균기온_1990_2024_통계청.xlsx')
전국평균기온.drop(range(0,6), inplace=True)
전국평균기온.drop('Unnamed: 1', axis=1, inplace=True)
전국평균기온.columns = 전국평균기온.loc[6]
전국평균기온.drop(range(6,25), inplace=True)
전국평균기온.drop(41, inplace=True)
전국평균기온.reset_index(drop=True, inplace=True)
전국평균기온
```

```python
사료급여량 = pd.read_excel('육우_두당_사육현황_2008_2023_통계청.xlsx')
사료급여량
```

```python
사료급여량.drop(range(0,43), inplace=True)
사료급여량.drop(range(44,58), inplace=True)
사료급여량=사료급여량.T
사료급여량.insert(0, 'Index', 사료급여량.index)
사료급여량.reset_index(drop=True, inplace=True)
사료급여량.columns = 사료급여량.loc[0]
사료급여량.drop(0, inplace=True)
사료급여량.columns=['년', '사료급여량 (Kg)']
사료급여량.reset_index(drop=True, inplace=True)
사료급여량
```

```python
전국평균기온['년'] = 전국평균기온['년'].astype(int)
사료급여량['년'] = 사료급여량['년'].astype(int)
temp_intake = pd.merge(전국평균기온, 사료급여량, on='년', how='inner')
temp_intake = temp_intake.astype({'년': int, '평균기온(℃)': float, '평균최저기온(℃)': float, '평균최고기온(℃)': float, '사료급여량 (Kg)': float})
temp_intake
```

```python
temp_intake.corr()
```

```python
plt.figure(figsize=(10,5))
plt.plot(temp_intake['년'], temp_intake['사료급여량 (Kg)'])

z = np.polyfit(temp_intake['년'], temp_intake['사료급여량 (Kg)'], 1)
p = np.poly1d(z)
plt.plot(temp_intake['년'], p(temp_intake['년']), 'k--')

plt.title('연도별 사료 급여량 변화(단위:kg)')
plt.xlim(2008,2023)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.grid()
plt.show()
```
