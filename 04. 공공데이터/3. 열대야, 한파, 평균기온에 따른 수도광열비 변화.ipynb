```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import matplotlib.ticker as ticker
import numpy as np
```

```python
열대야 = pd.read_csv('열대야일수(19902024).csv')
열대야.drop(columns=['6월', '7월', '8월', '9월', '순위'], inplace=True)
# 전국평균기온.drop(range(0,6), inplace=True)
# 전국평균기온.drop('Unnamed: 1', axis=1, inplace=True)
# 전국평균기온.columns = 전국평균기온.loc[6]
# 전국평균기온.drop(6, inplace=True)
# 전국평균기온.reset_index(drop=True, inplace=True)
열대야
```

```python
plt.figure(figsize=(10,5))
plt.bar(열대야['연도'], 열대야['연합계'], color='skyblue')

z = np.polyfit(열대야['연도'], 열대야['연합계'], 1)
p = np.poly1d(z)
plt.plot(열대야['연도'], p(열대야['연도']), '--', color='r')

plt.title('연도별 열대야 일수 변화')
# plt.xlim(,2023)
# plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
# plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.grid()
plt.show()
```

```python
전국평균기온 = pd.read_excel('전국평균기온_1990_2024_통계청.xlsx')
전국평균기온.drop(range(0,6), inplace=True)
전국평균기온.drop('Unnamed: 1', axis=1, inplace=True)
전국평균기온.columns = 전국평균기온.loc[6]
전국평균기온.drop(range(6,25), inplace=True)
전국평균기온.drop(41, inplace=True)
전국평균기온.reset_index(drop=True, inplace=True)
전국평균기온.columns=['연도', '평균기온(℃)','평균최저기온(℃)','평균최고기온(℃)']
전국평균기온 = 전국평균기온.astype({'연도': int, '평균기온(℃)': float, '평균최저기온(℃)': float, '평균최고기온(℃)': float})
전국평균기온
```

```python
plt.figure(figsize=(10,5))
plt.plot(전국평균기온['연도'], 전국평균기온['평균기온(℃)'])

z = np.polyfit(전국평균기온['연도'], 전국평균기온['평균기온(℃)'], 1)
p = np.poly1d(z)
plt.plot(전국평균기온['연도'], p(전국평균기온['연도']), '--', color='purple')

plt.title('연도별 평균 기온 변화(단위:℃)')
plt.xlim(2008,2023)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.grid()
plt.show()
```

```python
사육비 = pd.read_excel('육우_두당_사육비_2008_2023_통계청.xlsx')
사육비
```

```python
사육비.drop(range(0,7), inplace=True)
사육비.drop(range(8,37), inplace=True)
사육비=사육비.T
사육비.insert(0, 'Index', 사육비.index)
사육비.reset_index(drop=True, inplace=True)
사육비.columns = 사육비.loc[0]
사육비.drop(0, inplace=True)
사육비.columns=['연도', '수도광열비']
사육비.reset_index(drop=True, inplace=True)
사육비 = 사육비.astype({'연도': int, '수도광열비': int})
사육비
```

```python
plt.figure(figsize=(10,5))
plt.plot(사육비['연도'], 사육비['수도광열비'])

z = np.polyfit(사육비['연도'], 사육비['수도광열비'], 1)
p = np.poly1d(z)
plt.plot(사육비['연도'], p(사육비['연도']), '--', color='purple')

plt.title('연도별 수도광열비 변화(단위:원)')
plt.xlim(2008,2023)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.grid()
plt.show()
```

```python
total_x = pd.merge(열대야, 한파일수, on='연도', how='inner')
total_y = pd.merge(total_x, 전국평균기온,  on='연도', how='inner')
real_total = pd.merge(total_y, 사육비,  on='연도', how='inner')

real_total.columns=['연도', '열대야', '한파일수', '평균기온(℃)', '평균최저기온(℃)', '평균최고기온(℃)', '수도광열비']
real_total
```
